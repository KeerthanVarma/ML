```python
#3
import time
def finding_kth(l, k):
    # Make a copy of the list to avoid modifying the original list
    duplicate_l = l.copy()
    for i in range(k):
        max_index = 0
        for j in range(1, len(duplicate_l)):
            if duplicate_l[j] > duplicate_l[max_index]:
                max_index = j
        duplicate_l[-1], duplicate_l[max_index] = duplicate_l[max_index], duplicate_l[-1]
        required = duplicate_l.pop()
    return required 

def finding_median(l):
    n = len(l)
    if n % 2 == 0:
        median = (finding_kth(l, n // 2) + finding_kth(l.copy(), n // 2 + 1)) / 2
    else:

        median = finding_kth(l, n // 2 + 1)
    return median


def partition(arr, low, high):
    pivot = arr[high]
    i = low
    for j in range(low, high):
        if arr[j] < pivot:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
    arr[i], arr[high] = arr[high], arr[i]
    return i

def quickselect(arr, low, high, k):
    if low == high:
        return arr[low]
    
    pivot_index = partition(arr, low, high)
    
    if k == pivot_index:
        return arr[k]
    elif k < pivot_index:
        return quickselect(arr, low, pivot_index - 1, k)
    else:
        return quickselect(arr, pivot_index + 1, high, k)

def find_median_quickselect(arr):
    n = len(arr)
    if n % 2 == 1:
        return quickselect(arr, 0, n - 1, n // 2)
    else:
        lower_median = quickselect(arr, 0, n - 1, n // 2 - 1)
        upper_median = quickselect(arr.copy(), 0, n - 1, n // 2)
        return (lower_median + upper_median) / 2
def numpy_1(arr):
    med_numpy=np.median(arr)
    return med_numpy

def comp_methods(arr):
    start_time = time.time()
    med_inefficient = finding_median(arr.copy())
    inefficient_time = time.time() - start_time
    
    start_time = time.time()
    med_quick = find_median_quickselect(arr.copy())
    quick_time = time.time() - start_time
    
    start_time = time.time()
    med_numpy = numpy_1(arr)
    numpy_time = time.time() - start_time
    
    print(f"Inefficient Method: Median = {med_inefficient}, Time = {inefficient_time:.6f}s")
    print(f"Quickselect Method: Median = {med_quick}, Time = {quick_time:.6f}s")
    print(f"Numpy Method: Median = {med_numpy}, Time = {numpy_time:.6f}s")

comp_methods([1,3,5,2,4,7])

```

```python
#10
import numpy as np
import pandas as pd
np.random.seed(0)
Matrix=np.random.uniform(1, 2, size=(20, 5))
df=pd.DataFrame(Matrix, columns=["a", "b", "c", "d", "e"])
standard_deviation = df.std()
max_sd=standard_deviation.max()
max_sd_column=standard_deviation[standard_deviation==max_sd].index[0]
print(f"Column with the highest standard deviation: {max_sd_column}")
mean_r=df.mean(axis=1)
min_mean_row=mean_r[mean_r==mean_r.min()].index[0]
print(f"Row with the lowest mean: {min_mean_row}")
```
